```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.boxplot(x='sex',y='age',data=df,hue='survived')
```

```python
lst=["age"]
for i in lst:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    print(f"IQR: {iqr}")
    lowerbound=q1-1.5*iqr
    upperbound=q3+1.5*iqr
    print(f"lower Bound: {lowerbound}")
    print(f"upper Bound: {upperbound}")
```

```python
aged=df[['sex','age']].dropna()
aged.shape
data=aged[aged['sex']=='male']
data.info()
data
dataf=aged[aged['sex']=='female']

datam=data['age']
data=sorted(datam)
data
q1=np.percentile(data, 25)
q3=np.percentile(data,75)
iqr=q3-q1
lb=q1-(1.5 * iqr)
ub=q3 + (1.5 * iqr)
print('q1=', q1)
print('q3=',q3)
print('lower bound =',lb)
print('Upper bound =', ub)
out=[]
for i in data:
    if(i<lb):
        print(i)
    if(i>ub):
        print(i)

```
