```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
Ct=pd.read_csv("Ct.csv")
Ct
Ct['class Test Marks']
plt.boxplot(Ct['class Test Marks'])
outliers_zscore=[]
def detect_outliers_zscore(data):
    outliers_zscore=[]
    threshold=1
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if(np.abs(z_score)>threshold):
         outliers_zscore.append(i)
    return outliers_zscore
marks=[13,15,17,7,11,16,101,15,110,9,10]
marks_outliers = detect_outliers_zscore(marks)
print(marks_outliers)
from sklearn.preprocessing import MinMaxScaler
np.random.seed(42)
data=np.random.exponential(scale=10,size=1000)
df=pd.DataFrame({'original':data})
df['log_Trans']=np.log1p(df['original'])
df['sqrt_Trans']=np.sqrt(df['original'])
scaler=MinMaxScaler()
df['MinMax_Trans']=scaler.fit_transform(df[['original']])
df
fig,axes=plt.subplots(2,2,figsize=(12,8))
sns.histplot(df['original'],bins=30,kde=True,ax=axes[0,0])
sns.histplot(df['log_Trans'],bins=30,kde=True,ax=axes[0,1])
sns.histplot(df['sqrt_Trans'],bins=30,kde=True,ax=axes[1,0])
sns.histplot(df['MinMax_Trans'],bins=30,kde=True,ax=axes[1,1])
```
