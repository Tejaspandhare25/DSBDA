```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
Ct=pd.read_csv("Ct.csv")
Ct
Ct['class Test Marks']
plt.boxplot(Ct['class Test Marks'])
outliers_zscore=[]
def detect_outliers_zscore(data):
    outliers_zscore=[]
    threshold=1
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if(np.abs(z_score)>threshold):
         outliers_zscore.append(i)
    return outliers_zscore
marks=[13,15,17,7,11,16,101,15,110,9,10]
marks_outliers = detect_outliers_zscore(marks)
print(marks_outliers)
```
